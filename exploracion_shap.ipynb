```python
import numpy as np
import shap
import statsmodels.api as sm
from joblib import load
```

```python
modelo_logistico = load("model_2.pickle")
#modelo_logistico = sm.iolib.load_pickle('model.pickle')
X_test = np.load('X_test.npy')
```
